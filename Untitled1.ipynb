```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.feature_selection import RFE
import joblib
```

```python
df = pd.read_csv("nursery.csv")

df
```

```python
print(df.info())
print(df.describe())
```

```python
color = plt.get_cmap("tab20").colors
fig, axes = plt.subplots(4, 2, figsize=(16, 23))
categories = df.columns.tolist()[:8]
for i in range(4):
    for j in range(2):
        feature = categories[i * 2 + j]
        count = df[feature].value_counts()
        
        wedges, texts, autotexts = axes[i, j].pie(
            count, 
            autopct='%1.1f%%', 
            colors=color, 
            startangle=140, 
            wedgeprops={'edgecolor': 'black'}
        )
        
        axes[i, j].set_title(f"Distribution of {feature}", fontsize=18, weight='bold')
        
        for autotext in autotexts:
            autotext.set_color('white')
            autotext.set_fontsize(15)
            autotext.set_weight('bold')
        
        for text in texts:
            text.set_fontsize(12)
            text.set_weight('bold')
        
        axes[i, j].axis('equal')
        
        axes[i, j].legend(
            wedges, 
            count.index, 
            title="Categories", 
            loc="center left", 
            bbox_to_anchor=(1, 0, 0, 1),
            prop={"size": 16}
            )
plt.tight_layout()
plt.show()
```

```python
label_encoder = LabelEncoder()
df_encoded = df.copy()
features = df_encoded.select_dtypes(include="object").columns
for feature in features[:8]:
    df_encoded[feature] = label_encoder.fit_transform(df_encoded[feature])
df_encoded["children"] += 1
df_encoded
```

```python
X = df_encoded.iloc[:, :8]
y = df_encoded["Target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
logistic_reg = RFE(LogisticRegression(max_iter=1000), n_features_to_select=2).fit(X_train, y_train)
y_pred_logistic = logistic_reg.predict(X_test)
print("Logistic_reg", accuracy_score(y_pred_logistic, y_test))
```

```python
random_forest = RFE(RandomForestClassifier(), n_features_to_select=8).fit(X_train, y_train)
y_pred_forest = random_forest.predict(X_test)
print("Random_forest", accuracy_score(y_pred_forest, y_test))
```

```python
decision_tree = DecisionTreeClassifier().fit(X_train, y_train)
y_pred_decision = decision_tree.predict(X_test)
print("Decision_tree", accuracy_score(y_pred_decision, y_test))
```

```python
K_neighbor = KNeighborsClassifier().fit(X_train, y_train)
y_pred_kneighbor = K_neighbor.predict(X_test)
print("K_neighbor", accuracy_score(y_pred_kneighbor, y_test))
```

```python
print("Logistic_regression: ", classification_report(y_test, y_pred_logistic, zero_division=1))
print("Random_forest: ", classification_report(y_test, y_pred_forest, zero_division=1))
print("Decision_tree: ", classification_report(y_test, y_pred_decision, zero_division=1))
print("Kneighbors: ", classification_report(y_test, y_pred_kneighbor, zero_division=1))
```

```python

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
logistic = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_logistic))
random = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_forest))
decision = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_decision))
Kneighbor = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_kneighbor))
logistic.plot(cmap="Reds", ax=axes[0,0], values_format="d")
random.plot(cmap="Blues", ax=axes[0,1], values_format="d")
decision.plot(cmap="Greens", ax=axes[1,0], values_format="d")
Kneighbor.plot(cmap="Oranges", ax=axes[1,1], values_format="d")
axes[0,0].set_title("Logistic Regression")
axes[0,1].set_title("Random Forest")
axes[1,0].set_title("Decision Tree")
axes[1,1].set_title("K_Neighbors")
fig.set_edgecolor("black")
plt.tight_layout()
plt.show()
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}
grid_search = GridSearchCV(estimator=decision_tree, param_grid=param_grid, cv=3, n_jobs=1, scoring="accuracy")
grid_search.fit(X_train, y_train)
```

```python

best_params = grid_search.best_params_
Decision_tree_best = DecisionTreeClassifier(
    criterion=best_params["criterion"],
    max_depth=best_params["max_depth"],
    max_features=best_params["max_features"],
    min_samples_leaf=best_params["min_samples_leaf"],
    min_samples_split=best_params["min_samples_split"],
    splitter=best_params["splitter"]
)


Decision_tree_best.fit(X_train, y_train)
y_pred_best = Decision_tree_best.predict(X_test)
print(classification_report(y_test, y_pred_best, zero_division=1))
print(accuracy_score(y_test, y_pred_best))
```

```python
joblib.dump(Decision_tree_best,"Project_Model.pkl")
```
